```python
#import statements
import tableone
import pandas as pd
import HelperMethods
import dataframe_image as dfi

#input files
file_path = '/Users/divs/Box/1-BMC Smartguards/10-Processing and Visualization/8-TBI Pupillometry/Data/Intermediate Datasets/'
ground_truth = pd.read_excel('/Users/divs/Box/1-BMC Smartguards/10-Processing and Visualization/8-TBI Pupillometry/Data/Consolidated_Study_Cohort_Traits.xlsx')
ground_truth.Discharge_Disposition = ground_truth.Discharge_Disposition.replace({"Skilled Nursing Facility":"Rehab Facility"})
print(ground_truth.columns)
```

# Full Cohort

```python
# columns to summarize
columns = [ 'tbi_severity', 'AGE', 'SEX', 'RACE', 'GCS', 'Mechanism_Injury','Crani_Surgery', 'Bolt','EVD', 'Marshall', 'Rotterdam', \
           'Time_to_most_severe_CT', 'n_obs',\
  'time_to_first_obs', 'average_npi', 'lower_npi', 'npi_diff', 'size_diff', 'average_npi_0_removed', 'lower_npi_0_removed', \
 'any_incidence', 'poor_npi_incidence','npi_diff_incidence', 'size_diff_incidence', 'uni_any_incidence', 'uni_stage_incidence',\
       'stage1u_incidence', 'stage2u_incidence', 'stage3u_incidence','stage4u_incidence', \
       'bi_incidence', 'stage1b_incidence','stage2b_incidence', 'stage3b_incidence', \
        'npi_diff_size_diff_incidence', 'size_diff_poor_npi_incidence',\
  'any_burden', 'poor_npi_burden', 'npi_diff_burden', 'size_diff_burden','uni_any_burden',\
      'uni_stage_burden', 'stage1u_burden', 'stage2u_burden', 'stage3u_burden', 'stage4u_burden', \
        'bi_burden', 'stage1b_burden', 'stage2b_burden','stage3b_burden',\
      'npi_diff_size_diff_burden','size_diff_poor_npi_burden',\
'Discharge_Disposition','Unfavorable_Outcome', 'Deceased', 'Orientedx3', 'Awake_Alert'\
]

# columns containing categorical variables
categorical = ['SEX', 'RACE', 'Mechanism_Injury', 'Discharge_Disposition', 'Orientedx3', 'Awake_Alert', \
               'Deceased', 'Unfavorable_Outcome', 'Crani_Surgery', 'Bolt', 'EVD',\
                'any_incidence', 'poor_npi_incidence','npi_diff_incidence', 'size_diff_incidence', 'uni_any_incidence',\
       'uni_stage_incidence', 'stage1u_incidence', 'stage2u_incidence', 'stage3u_incidence','stage4u_incidence', \
       'bi_incidence', 'stage1b_incidence','stage2b_incidence', 'stage3b_incidence', \
        'npi_diff_size_diff_incidence', 'size_diff_poor_npi_incidence']

# non-normal variables
nonnormal = ['AGE', 'GCS', \
             'n_obs','time_to_first_obs', 'average_npi', 'lower_npi', 'npi_diff', 'size_diff', 'average_npi_0_removed', 'lower_npi_0_removed',\
              'any_burden', 'poor_npi_burden', 'npi_diff_burden', 'size_diff_burden','uni_any_burden',\
      'uni_stage_burden', 'stage1u_burden', 'stage2u_burden', 'stage3u_burden', 'stage4u_burden', \
        'bi_burden', 'stage1b_burden', 'stage2b_burden','stage3b_burden',\
      'npi_diff_size_diff_burden','size_diff_poor_npi_burden']

# limit the binary variable "death" to a single row
limit = {"SEX": 2, "Mechanism_Injury":2, 'Orientedx3':1, 'Awake_Alert':1, 'Crani_Surgery':2, 'Bolt':2, 'EVD':2, 'Unfavorable_Outcome':2}
limit_vars = ['stage1u_incidence',\
'stage2u_incidence', 'stage3u_incidence', 'stage3u_v2_incidence', 'stage4u_incidence', 'stage1b_incidence',\
'stage2b_incidence', 'uni_incidence', 'bi_incidence', 'any_incidence', 'npi_diff_incidence', 'size_diff_incidence']

# set decimal places for age to 0
decimals = dict()
for c in columns:
    decimals[c] = 0
decimals.pop('tbi_severity')
for col in ['average_npi', 'lower_npi', 'npi_diff', 'size_diff', 'average_npi_0_removed', 'lower_npi_0_removed',  'Marshall', 'Rotterdam']:
  decimals[col] = 1

# optionally, a categorical variable for stratification
groupby = ['tbi_severity']

#create table_one using TableOne package
table1 = tableone.TableOne(ground_truth, columns=columns, categorical=categorical, missing=False, groupby=groupby,
                  nonnormal=nonnormal, label_suffix=True, limit=limit,
                  decimals=decimals,  pval=True)

table1
```
